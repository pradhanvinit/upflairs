```python
def showing(im):
    plt.imshow(im[:,:,::-1])
    plt.show()
def imshow(im):
    cv2.imshow('preview',im)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
showing(img1)
showing(img2)
```

```python
imshow(img1)
imshow(img2)
```
